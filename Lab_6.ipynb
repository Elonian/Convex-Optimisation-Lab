```
import numpy as np
import cvxpy as cp

profit_ipad = 10
profit_iphone = 6
profit_ipod = 4

max_ipad_inday = 3000
max_iphone_inday = 5000
max_ipod_inday = 6000

ipad_storg_perunit = 210.0/1000 
iphone_storg_perunit = 45.0/1000
ipod_storg_perunit = 40.0/1000


```

```
x1 = cp.Variable(1) ## ipad work fraction
x2 = cp.Variable(1) ## iphone work fraction
x3 = cp.Variable(1) ## ipod work fraction 

# Assuming each day can be split into some three parts which can be float value 
# All the five days follow the same shedule as the first day
# No of items produce is dependent on time devoted for the particular product

cost = 5.0 * (profit_ipad * max_ipad_inday * x1 + profit_iphone * max_iphone_inday * x2 + profit_ipod * max_ipod_inday * x3)

objective = cp.Maximize(cost)
constraint = [x1 + x2 + x3 <= 1,
              5 * ipad_storg_perunit * max_ipad_inday * x1 + 5 * iphone_storg_perunit * max_iphone_inday * x2 + 5 * ipod_storg_perunit * max_ipod_inday * x3 <= 6000,
              5 * max_ipod_inday * x3 >= 5000,
              5 * max_ipad_inday * x1 >= 4000,
              5 * max_ipod_inday * x3 <= 10000,
              5 * max_ipad_inday * x1 <= 8000,
              5 * max_iphone_inday * x2 <= 15000,
              x1 >= 0,
              x2 >= 0,
              x3 >= 0,]
prob = cp.Problem(objective, constraint)
prob.solve()

print("Profit: ",prob.value)
print("ipad work(fraction) per day ", x1.value)
print("iphone work(fraction) per day ", x2.value)
print("ipod work(fraction) per day ", x3.value)
```

```
y1 = cp.Variable(1) ## ipad covers per week
y2 = cp.Variable(1) ## iphone covers per week
y3 = cp.Variable(1) ## ipod covers per week

# Assuming each day can be split into some three parts which can be float value 
# All the three days follow the same shedule as the first day
# No of items produce is dependent on time devoted for the particular product

cost = profit_ipad * y1 + profit_iphone * y2 + profit_ipod * y3

objective = cp.Maximize(cost)
constraint = [y1/max_ipad_inday + y2/max_iphone_inday + y3/max_ipod_inday <= 5,
              ipad_storg_perunit * y1 + iphone_storg_perunit * y2 + ipod_storg_perunit * y3 <= 6000,
              y3 >= 5000,
              y1 >= 4000,
              y3 <= 10000,
              y1 <= 8000,
              y2 <= 15000,
              y1 >= 0,
              y2 >= 0,
              y3 >= 0,]
prob = cp.Problem(objective, constraint)
prob.solve()

print("Profit: ",prob.value)
print("ipad covers per week ", y1.value)
print("iphone covers per week ", y2.value)
print("ipod covers per week ", y3.value)
```

```
z1 = cp.Variable(1) ## Total production hours per week
z2 = cp.Variable(1) ## iphone production hours per week
z3 = cp.Variable(1) ## ipod production hours per week

time = 8.0 # No of production hrs per day

# Assuming each day can be split into some three parts which can be float value 
# All the three days follow the same shedule as the first day
# No of items produce is dependent on time devoted for the particular product

cost = profit_ipad * (max_ipad_inday / time) * (z1 - z2 - z3) + profit_iphone * (max_iphone_inday / time) * z2 + profit_ipod * (max_ipod_inday / time) * z3

objective = cp.Maximize(cost)
constraint = [ z1/time + z2/time + z3/time <= 8,
              ipad_storg_perunit * (max_ipad_inday/time) * (z1 - z2 - z3) + iphone_storg_perunit * (max_iphone_inday/time) * z2 + ipod_storg_perunit * (max_ipod_inday/time) * z3 <= 6000,
              (max_ipod_inday/time) * z3 >= 5000,
              (max_ipad_inday/time) * (z1 - z2 - z3) >= 4000,
              (max_ipod_inday/time) * z3 <= 10000,
              (max_ipad_inday/time) * (z1 - z2 - z3) <= 8000,
              (max_iphone_inday/time) * z2 <= 15000,
              z1 <= time * 5,
              z2 <= time * 5,
              z3 <= time * 5,
              z1 >= 0,
              z2 >= 0,
              z3 >= 0,]
prob = cp.Problem(objective, constraint)
prob.solve()

print("Profit: ",prob.value)
print("ipad production hrs per week ", z1.value - z2.value - z3.value)
print("iphone production hrs per week ", z2.value)
print("ipod production hrs per week ", z3.value)
print("Total production hrs: ",z1.value)
```
