```
x = cp.Variable((m,1)) ## Variables
b = cp.Variable((n,1))
w = cp.Variable((n,1))

objective = cp.Minimize(b[0]+np.ones(m)@x) # Total investment = b1(bank balance) + 1Transpose*x(investements)
constraints = [b >= 0, w + P@x >= e, x >= 0, w >= 0] 
for i in range(1,n,1):
    constraints += [b[i] == ((1+rho)*b[i-1]) - w[i-1]] ## t runs from 0 to n-1

prob = cp.Problem(objective, constraints)
prob.solve()

print("Min inital Investment: ",objective.value)
```
