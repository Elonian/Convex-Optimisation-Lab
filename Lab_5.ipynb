```
import numpy as np
import cvxpy as cp

def total_cost(x,Cost):
    total = 0
    for i in range(len(Cost)):
        for j in range(len(Cost[0])):
            total += x[i][j]* Cost[i][j]
    return total

Cost = [[8, 6, 10, 9, 8],
        [9, 12, 13, 7, 5],
        [14, 9, 16, 5, 2]]
demand = [45, 20, 30, 30, 10]
supply = [40, 50, 45]

x = cp.Variable((len(Cost),len(Cost[0])))

objective = cp.Minimize(total_cost(x,Cost))
constraint = [x >= 0]
for i in range(len(Cost)):
    constraint += [sum(x[i]) <= supply[i]]
for i in range(len(Cost[0])):
    constraint += [sum(x[:,i]) >= demand[i]]

prob = cp.Problem(objective, constraint)
prob.solve()
print("Optimal Cost:",objective.value)
print("Transport strategy:",x.value)
```
