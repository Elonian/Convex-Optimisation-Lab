Diet Problem 

Calories between 2000 to 2250

Vitamin A between 5000 to 50000

Minimise the cost


```

# The diet problem 

import cvxpy as cp
import numpy as np

cost = np.matrix( [0.18,0.23,0.05]) #Cost of corn, milk, Wheat Bread
x = cp.Variable(3)                  #No of items 
objective = cp.Minimize(cost*x)     #Objective to minimise over all cost

constraints = [x[0]*107+x[1]*500+x[2]*0 >= 5000,x[0]*107+x[1]*500+x[2]*0 <= 50000,x[0]*72+x[1]*121+x[2]*65 >= 2000,x[0]*72+x[1]*121+x[2]*65 <= 2250,x[0] <= 10,x[1]<= 10,x[2]<= 10,x[0] >= 0,x[1]>= 0,x[2]>= 0]

prob = cp.Problem(objective, constraints)
result = prob.solve()
print("No of items: ",x.value)
print("Minimum cost: ", objective.value)
print("Rounded Value: ",round(objective.value,5))
```
